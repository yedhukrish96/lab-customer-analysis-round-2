```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
#here starts round 5
#Normalising
y = numerical['Total Claim Amount']
X = numerical.drop(['Total Claim Amount'], axis=1)

#Normalising-
transformer = MinMaxScaler().fit(X)
x_normalised= transformer.transform(X)
print(x_normalised.shape)
x_normalised=pd.DataFrame(x_normalised, columns=X.columns)
```

```python
#Here starts round 6
catagorical.head()
```

```python
catagorical= catagorical.drop([ 'Effective To Date'], axis=1)
catagorical.head()
```

```python
X_cat = catagorical.select_dtypes(include = object)
X_cat.head()
```

```python
X_cat['Coverage'].value_counts()
```

```python
df1 = pd.DataFrame(data=X_cat, columns=['Coverage'])
display(df1.value_counts())
display(df1)


```

```python
#one-hot encoding
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder().fit(df1)
encoder = OneHotEncoder(drop='first').fit(df1)
print(encoder.categories_)


encoded = encoder.transform(df1).toarray()
print(encoded)
onehot_encoded = pd.DataFrame(encoded,columns=['Extended', 'Premium'])
onehot_encoded.head(20)
```

```python
#Concatenating the two dataframes
X= pd.concat([x_normalised,onehot_encoded],axis=1)
display(X)
```

```python
X.head()
```

```python
y = numerical['Total Claim Amount']
y.head()
```

```python
#Doing the train-test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# check for missing values in the dataset
print(X_train.isna().sum())

# fill missing values with the median of each column
X_train.fillna(X_train.median(), inplace=True)

# check for missing values again to make sure they have been filled
print(X_train.isna().sum())
```

```python
#Doing the same to other data as it showed error for me
print(X_test.isna().sum())

X_test.fillna(X_test.median(), inplace=True)
print(X_test.isna().sum())
```

```python
#Applying Linear Regression
from sklearn import linear_model

lm = linear_model.LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Model Validation-
#Predicting the R2 score-

from sklearn.metrics import r2_score
predictions = lm.predict(X_train)
r2_score(y_train, predictions)
```

```python
# Applying predictions on the test part-

predictions_test = lm.predict(X_test)
r2_score(y_test, predictions_test)
```

```python
# Finding the mean squared error-
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,predictions_test)
mse
```

```python
# Finding the root of the mean squared error-
rmse = np.sqrt(mean_squared_error(y_test,predictions_test))
rmse
```
